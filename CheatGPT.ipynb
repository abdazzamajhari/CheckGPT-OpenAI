```python
import os
import torch
from transformers import RobertaTokenizer, RobertaForSequenceClassification
from langdetect import detect
from translate import Translator

model_name = 'roberta-base-openai-detector'

# Periksa apakah model sudah diunduh
if not os.path.exists(model_name):
    print(f"Model '{model_name}' belum diunduh. Mulai mengunduh...")
    tokenizer = RobertaTokenizer.from_pretrained(model_name)
    model = RobertaForSequenceClassification.from_pretrained(model_name)
    print(f"Model '{model_name}' telah diunduh.")
else:
    print(f"Model '{model_name}' sudah ada.")

# Inisialisasi translator Microsoft Translator
translator = Translator(from_lang='id', to_lang='en')

# Masukkan teks yang ingin dideteksi
teks = input("Masukkan teks: ")

# Periksa bahasa input
detected_language = detect(teks)

# Terjemahkan jika bahasa input adalah bahasa Indonesia
if detected_language != 'en':
    translation = translator.translate(teks)
    teks = translation

# Tokenisasi teks
tokenizer = RobertaTokenizer.from_pretrained(model_name)
tokens = tokenizer.encode_plus(teks, add_special_tokens=True, max_length=512, truncation=True, padding='longest', return_tensors='pt')

# Prediksi menggunakan model
model = RobertaForSequenceClassification.from_pretrained(model_name)
outputs = model(**tokens)
predictions = torch.softmax(outputs.logits, dim=1).tolist()[0]

# Deteksi probabilitas AI-generated
ai_generated_prob = predictions[1] * 100

# Tampilkan hasil deteksi
if ai_generated_prob > 50:
    print("Tulisan ini kemungkinan besar dihasilkan oleh AI.")
    print(teks)
    print(f"Probabilitas AI-generated: {ai_generated_prob:.2f}%")
else:
    print("Tulisan ini kemungkinan bukan hasil dari AI.")
    print(teks)
    print(f"Probabilitas AI-generated: {ai_generated_prob:.2f}%")
```
